```python
import yfinance as yf
import numpy
import pandas as pd
```

```python
stockTicker = "F"

autom = yf.Ticker("^QAUTO")
sp500 = yf.Ticker("^GSPC")
stock = yf.Ticker(stockTicker)

#Data of price every day of last 10 years
dataAutom = autom.history(period='10y',interval='1d')
dataSp500 = sp500.history(period='10y',interval='1d')
dataStock = stock.history(period='10y',interval='1d')
```

```python
infoStock = stock.info
display(infoStock.keys())
```

```python
#Various ratios and pieces of information can be called like this:
display(infoStock['longName'])
display(infoStock['industry'])
display(infoStock['beta'])
display(infoStock['forwardPE'])
```

```python
display(dataStock) #calculate avg price per day
lenStock = len(dataStock)
avgPrice = []
for i in range(lenStock):
    avg = (dataStock['High'][i]+dataStock['Low'][i])/2
    avgPrice.append(avg)
avgPriceDf = pd.DataFrame({'avgPrice':avgPrice},index=dataStock.index.values)
display(avgPriceDf)
```

```python
options = stock.option_chain()  #option_chain(date='2020-07-24') returns all option with expiration date 2020-07-24
calls = options.calls
display(calls)
puts = options.puts
display(puts)
```
