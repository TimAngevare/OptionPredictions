```python
import yfinance as yf
import numpy
import pandas as pd
```

```python
stockTicker = "F"

autom = yf.Ticker("^QAUTO")
sp500 = yf.Ticker("^GSPC")
stock = yf.Ticker(stockTicker)

#Data of price every minute from 10th of October to 17th of October
dataAutom = autom.history(interval='1m', start='2022-10-10', end='2022-10-17')
dataSp500 = sp500.history(interval='1m', start='2022-10-10', end='2022-10-17')
dataStock = stock.history(interval='1m', start='2022-10-10', end='2022-10-17')
```

```python
infoStock = stock.info
display(infoStock.keys())
```

```python
display(infoStock['longName'])
display(infoStock['industry'])
display(infoStock['beta'])
display(infoStock['forwardPE'])
```

```python
display(dataStock)
lenStock = len(dataStock)
avgPrice = []
for i in range(lenStock):
    avg = (dataStock['High'][i]+dataStock['Low'][i])/2
    avgPrice.append(avg)
avgPriceDf = pd.DataFrame({'avgPrice':avgPrice},index=dataStock.index.values)
display(avgPriceDf)
```
